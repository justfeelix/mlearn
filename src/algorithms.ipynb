## Phase 9: Testing, Benchmarking, Reflection
Goal:
- write repeatable checks
- measure performance trends
- document lessons learned

```python
# TODO: tiny benchmark template
import time

def benchmark(fn, data, repeats=5):
    times = []
    for _ in range(repeats):
        t0 = time.perf_counter()
        fn(data.copy() if hasattr(data, 'copy') else data)
        times.append(time.perf_counter() - t0)
    return min(times), sum(times) / len(times)
```
